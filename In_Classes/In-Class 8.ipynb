```python
# (a)
import boto3
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression, Perceptron
from sklearn.metrics import accuracy_score

# Defining s3 bucket
s3 = boto3.resource('s3')
bucket_name = 'gabrielferreira-data-455-bucket'
bucket = s3.Bucket(bucket_name)

# Defining the file to be read from s3 bucket
key_file = 'framingham.csv'

bucket_object = bucket.Object(key_file)
file_object = bucket_object.get()
file_content_stream = file_object.get('Body')

# Reading csv
heart = pd.read_csv(file_content_stream)

# (b)
heart = heart.dropna()
heart.head(1)
```

```python
# Defining the input and target variables
X = heart[['age', 'currentSmoker', 'totChol', 'BMI', 'heartRate', 'glucose']]
Y = heart['TenYearCHD']

# L2 Normalization
def l2_normalization(X):
    
    x_mean = np.mean(X)
    l2 = np.sqrt(sum(X**2))
    
    return (X - x_mean) / l2
```

```python
from sklearn.model_selection import train_test_split

# Creatinmg list to store the accuracy socres
md1_results = []
md2_results = []

for i in range(0,100):
     
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)

    X_train = X_train.apply(l2_normalization, axis = 1)
    X_test = X_test.apply(l2_normalization, axis = 1)
    
    # Building Logistic Regression Model
    md1 = LogisticRegression().fit(X_train, Y_train)

    # Predicting on the test dataset
    pred1 = md1.predict_proba(X_test)[:,1]
    pred1 = np.where(pred1 < 0.25, 0, 1)

    # Computing the accuracy of the model
    md1_results.append(accuracy_score(Y_test, pred1))
    
    ## Model 2
    md2 = Perceptron(penalty = 'l2', alpha = 0.01).fit(X_train, Y_train)
    
    # Predicting on the dataset
    pred2 = md2.predict(X_test)

    # Computing the accuracy
    md2_results.append(accuracy_score(Y_test, pred2))
```

```python
md_results = pd.DataFrame({"Model_1" : md1_results, "Model_2": md2_results})
md_results.head()
```

```python
# Visualizing model 1
import matplotlib.pyplot as plt

## Plotting iteration results
fig = plt.figure(figsize = (12, 8))

iterations = range(0, 100)
plt.plot(iterations, md1_results, marker = 'o', color = 'blue',
         label = 'First Model')
plt.plot(iterations, md2_results, marker = 'o', color = 'orange',
        label = 'Second Model')
plt.xlabel('Number of Iteration')
plt.ylabel('Accuracy Score')
plt.legend()
plt.grid()
plt.show()
```
