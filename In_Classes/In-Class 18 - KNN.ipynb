```python
## Question 
# (a)
import boto3
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import NearestNeighbors

# Defining s3 bucket
s3 = boto3.resource('s3')
bucket_name = 'gabrielferreira-data-455-bucket'
bucket = s3.Bucket(bucket_name)

# Defining the file to be read from s3 bucket
key_file = 'customer_offers.csv'
key_file_1 = 'offer_info.csv'

bucket_object = bucket.Object(key_file)
bucket_object_1 = bucket.Object(key_file_1)

file_object = bucket_object.get()
file_object_1 = bucket_object_1.get()

file_content_stream = file_object.get('Body')
file_content_stream_1 = file_object_1.get('Body')

# Reading csv
offers = pd.read_csv(file_content_stream)
offer_info = pd.read_csv(file_content_stream_1)

offers.head()
```

```python
offers.describe()
```

```python
# Dropping customer_name
offers = offers.drop(columns = ['customer_name'], axis = 1)
offers.head(1)
```

#### We do not need to standardize the data because it is already in 0-1 scale

```python
## Fitting the 10-nearest neighbor model to the data
knn_md = NearestNeighbors(n_neighbors = 10, algorithm = 'brute').fit(offers)

## Defining the observation of interst (15)
x_int = offers.loc[14]

## Extracting the 10-nearest neighbors of observation
knn_md.kneighbors([x_int])
```

```python
# Taking a look at the neighbors
offers.loc[[14, 63, 96, 3, 18, 32, 58, 36, 42, 46]].apply(sum, axis = 0)
```

7 out of the 10 neighbors like the 30th wine, which is Malbac. Also, customers with these features are more likely to buy the items 8, 11, 7, 18, 22, 12, 29 respectively.
